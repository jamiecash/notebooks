# The worlds most annoying hello world programme.
Prints HELLO WORLD using character positions contained in list

```python
letters = "ABCDEFGHIJKLMNOPQRSTUVWXYZ "
indexes = [8, 5, 12, 12, 15, 27, 23, 15, 18, 12, 4]

for index in indexes:
    letter = letters[index-1]
    print(f"{letter}")
```
