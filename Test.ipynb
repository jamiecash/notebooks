## Test that home directory is on NAS, not local
Total space should be really large. Over 10 TB. If this was local, the total space would be the size of the os install volume

```python
import os
import shutil

# Get disk usage for home
total, used, free = shutil.disk_usage(".")

# In TB
total_tb = total / 1024 / 1024 / 1024 / 1024

# It should be over 10TB
"OK: Home dir is on NAS" if total_tb > 10 else "ERROR: Home dir is local"
```
